```python
directory="altman_z_all.xlsx"
import pandas as pd
import numpy as np
```

```python
#data=pd.read_excel(directory,encoding="latin-1", skiprows = [0,1,2,4], na_values = "#N/A N/A", sheet_name = "Sheet1", index_col = 0, header = None).dropna(axis = 1, how = "all").T
```

```python


cols = pd.read_excel(directory, encoding="latin-1", skiprows = [0,1,2,4,5], na_values = ["#N/A N/A", "#N/A Review"], sheet_name = "Sheet1", header=None, nrows=1, index_col = 0).values[0]
data_r = pd.read_excel(directory, encoding="latin-1", skiprows = [0,1,2,4,5], na_values = ["#N/A N/A", "#N/A Review"], sheet_name = "Sheet1", index_col = 0).astype(np.float64)
data_r.columns = cols
data=data_r.dropna(axis=1,how="all")
data = data.loc[:,~data.columns.duplicated()]
dataT=data.T
display(data)
```

```python

data_lower = data.where(data<=1.8)
data_lower = data_lower.dropna(axis=1,how="all")
#data_lower.to_csv("data_lower.csv")

columns_to_keep=list(data_lower.columns)
data_filtered=data.filter(columns_to_keep,axis=1)
display(data_filtered)
```

```python
data_test = data_lower
```

```python
data_lower
```

```python
data_test = data_lower.notna().applymap(lambda x: 1 if x else 0)
data_test
```

```python
data_test_applied = data_test.apply(lambda y: y * (y.groupby((y != y.shift()).cumsum()).cumcount() + 1))
```

```python
#data_test_applied.to_csv("counted_list.csv")
```

```python
#The longest these companies have stayed below altman-z-score of 1.8
longest_under = data_test_applied.apply(lambda x: x.max()).sort_values(ascending = False).to_frame()
longest_under 
```

```python
#prototype for seeing the length of the consecutive quarters below 1.8

#data_first_time = data_test_applied["SO UN Equity"].value_counts().sort_values().to_frame()

data_first_time = data_test_applied.apply(lambda x: x.value_counts().sort_values())
display(data_first_time)
```

```python
#prototype: number of times the company first fell below 1.8
data_falls = len(data_test_applied[data_test_applied["BA UN Equity"] == 1])
data_falls
```
